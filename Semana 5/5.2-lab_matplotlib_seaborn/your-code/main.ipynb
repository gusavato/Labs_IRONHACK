```python
# your code here

sns.histplot(titanic.Age, bins = 100);

plt.axvline(titanic.Age.mean(),
            c='r',
            linewidth=3.,
            linestyle='--',
            label='media de edad')

plt.axvline(titanic.Age.median(),
            c='orange',
            linewidth=3.,
            linestyle='--',
            label='mediana de edad')

plt.axvline(titanic.Age.mode()[0],
            c='black',
            linewidth=3.,
            linestyle='--',
            label='moda de edad')

plt.legend();
```
